```python
from log_utils import get_all_log_paths, import_log, get_total_flight_time
```

```python
search_path = 'C:/Users/aeagr/University of Southampton/CASCADE - Documents/Flight logs repository'
```

```python
file_list = get_all_log_paths(search_root=search_path)
file_list
```

```python
durations = {}
for (ii,fn) in enumerate(file_list):
    this_df = import_log(fn, include_types=['PARM'])
    this_csv = f'cascade/result{ii:000d}.csv'
    try:
        this_dur = get_total_flight_time(this_df)
    except ValueError:
        this_dur = 0
    print(f'Flight duration {this_dur} from file {fn}')
    durations[fn] = this_dur
    this_df.to_csv(this_csv)
```
